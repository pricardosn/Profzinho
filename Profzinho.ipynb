```python
import google.generativeai as genai
from google.colab import userdata
API_KEY = userdata.get("secret_key")

# Inicializa a API do Gemini
genai.configure(api_key=API_KEY)

def coletar_informacoes_aluno():
    """Coleta informações relevantes sobre o aluno."""
    nome = input("Digite o nome do aluno: ")
    idade = int(input("Digite a idade do aluno: "))
    sexo = input("Digite o sexo do aluno (masculino/feminino): ").lower()
    serie = input("Digite a série do aluno: ")
    return nome, idade, sexo, serie

def definir_objetivos_estudo():
    """Define os objetivos de estudo do aluno."""
    conteudo = input("Digite a matéria a ser estudada: ")
    objetivos = input("Digite o conteúdo a ser estudado: ")
    return conteudo, objetivos

def gerar_roteiro_estudo(nome, idade, sexo, serie, conteudo, objetivos):
    """Gera um roteiro de estudo personalizado usando o Gemini."""
    prompt = f"""
    Crie um roteiro de estudo detalhado para {nome}, um aluno de {idade} anos, {sexo},
    cursando a {serie} série. O conteúdo a ser estudado é {conteudo}, com o objetivo de {objetivos}.

    Considere os seguintes aspectos ao elaborar o roteiro:
    * **Plano de Estudo Semanal:** Divida o conteúdo em tópicos semanais.
    * **Atividades Diversificadas:** Inclua atividades variadas como leitura de livros, exercícios, vídeos explicativos e jogos educativos.
    * **Recursos de Apoio:** Sugira livros, sites, aplicativos e outras ferramentas úteis.
    * **Tempo de Estudo:** Defina um tempo de estudo diário e semanal adequado à idade e rotina do aluno.
    * **Dicas de Aprendizagem:** Inclua dicas de estudo personalizadas, considerando o sexo e a série do aluno.
    * **Organização e Motivação:** Incentive o aluno a se organizar, criar um ambiente de estudo adequado e manter a motivação.

    O roteiro deve ser apresentado de forma clara, organizada e fácil de ser seguida por um aluno da {serie} série.
    """
    model = genai.GenerativeModel(model_name="gemini-1.0-pro")

    resposta = model.generate_content(prompt)
    print(resposta.text)

    return f"**Roteiro de Estudo para {nome} **\n\n{resposta.text}"

def salvar_roteiro(roteiro):
    """Salva o roteiro em um arquivo de texto."""
    with open("roteiro_estudo.txt", "w", encoding='utf-8') as arquivo:
        arquivo.write(roteiro)

if __name__ == "__main__":
    nome, idade, sexo, serie = coletar_informacoes_aluno()
    conteudo, objetivos = definir_objetivos_estudo()
    roteiro = gerar_roteiro_estudo(nome, idade, sexo, serie, conteudo, objetivos)
    print(roteiro)
    salvar_roteiro(roteiro)
    print("Roteiro de estudo salvo em 'roteiro_estudo.txt'")
```
